```python
result = generate_recommendation(
    age_months=18,
    muac=11.2,
    waz=-3.1,
    haz=-2.4,
    whz=-2.8,
    underweight=1,
    stunting=1,
    wasting=1
)

print(result)
```

```python
def display_result(result):

    print("=" * 60)
    print("PEDIATRIC MALNUTRITION ASSESSMENT")
    print("=" * 60)

    print("\nPredictions")
    print("-" * 60)
    for key, value in result["Prediction"].items():
        print(f"{key:<20}: {'Positive' if value else 'Negative'}")

    print("\nAssessment")
    print("-" * 60)
    for key, value in result["Assessment"].items():
        print(f"{key:<20}: {value}")

    print("\nNutrition Recommendations")
    print("-" * 60)
    for i, rec in enumerate(result["Recommendations"], 1):
        print(f"{i}. {rec}")

    if result["Alerts"]:
        print("\nMedical Alerts")
        print("-" * 60)
        for alert in result["Alerts"]:
            print(f"⚠️ {alert}")
```

```python
display_result(result)
```
