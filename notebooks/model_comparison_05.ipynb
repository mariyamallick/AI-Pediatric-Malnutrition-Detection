```python
import pandas as pd
from pathlib import Path

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)
```

```python
DATA_PATH = Path("../data/processed/synthetic_feature_engineered.csv")

df = pd.read_csv(DATA_PATH)
```

```python
features = [
    "age_months",
    "sex",
    "weight_kg",
    "height_cm",
    "muac_cm",
    "waz",
    "haz",
    "whz",
    "bmi"
]

X = df[features]
```

```python
y = df["underweight_status"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),

    "Decision Tree": DecisionTreeClassifier(random_state=42),

    "Random Forest": RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )
}
```

```python
results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)

    precision = precision_score(
        y_test,
        predictions,
        average="weighted"
    )

    recall = recall_score(
        y_test,
        predictions,
        average="weighted"
    )

    f1 = f1_score(
        y_test,
        predictions,
        average="weighted"
    )

    results.append([
        name,
        accuracy,
        precision,
        recall,
        f1
    ])
```

```python
results_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "Accuracy",
        "Precision",
        "Recall",
        "F1 Score"
    ]
)

results_df
```

```python
best_model = results_df.sort_values(
    by="Accuracy",
    ascending=False
)

best_model
```

```python
import joblib
from pathlib import Path

best_model_name = best_model.iloc[0]["Model"]

selected_model = models[best_model_name]

model_dir = Path("../models")
model_dir.mkdir(parents=True, exist_ok=True)

joblib.dump(
    selected_model,
    model_dir / "best_model.pkl"
)

print(f"{best_model_name} saved successfully!")
```

```python
results_df.to_csv(
    "../reports/model_comparison.csv",
    index=False
)
```
